## 4.1.2. [<img src="https://raw.githubusercontent.com/FortAwesome/Font-Awesome/6.x/svgs/solid/arrow-rotate-left.svg" width="20" height="20" style="filter: invert(21%) sepia(96%) saturate(3035%) hue-rotate(204deg) brightness(99%) contrast(103%);">](#00)

Для заданной таблицы значений функции (следа $y(x) = 4x^3 + 1$) построить линейный, квадратичный и кубический интерполяционные многочлены для оценки $y(x_0)$. Сравнить погрешности интерполяции с величиной остаточных многочленов.

| x | $0.00$ | $1.00$ | $2.00$ | $3.00$ | $4.00$ | $5.00$ | $6.00$ |
| --- | --- | --- | --- | --- | --- | --- | --- |
| y | $1.00$ | $5.00$ | $33.00$ | $109.00$ | $257.00$ | $501.00$ | $865.00$ |

\begin{equation*}
    \text{а) } x_0 = 0.5; \qquad \qquad \qquad \text{б) } x_0 = 2.5.
\end{equation*}

### а)

#### Линейная интерполяция:

Берём узлы $x=0$ и $x=1$:

\begin{equation*}
    f(0)=1,\quad f(1)=5.
\end{equation*}

Линейный интерполяционный многочлен:

\begin{equation*}
    P_1(x) = f(0) \frac{x-1}{0-1} + f(1) \frac{x-0}{1-0} = 1 \cdot \frac{x-1}{-1} + 5x = -(x-1) + 5x = 4x + 1.
\end{equation*}

В точке $x_0 = 0.5$:

\begin{equation*}
    P_1(0.5) = 4 \cdot0.5 + 1 = 2 + 1 = 3.
\end{equation*}

Точное значение:

\begin{equation*}
    f(0.5) = 4(0.5^3) + 1 = 4 \cdot 0.125 + 1 = 0.5 + 1 = 1.5.
\end{equation*}

Ошибка:

\begin{equation*}
    \sigma_1 = |f(0.5) - P_1(0.5)| = |1.5 - 3| = 1.5.
\end{equation*}

Остаточный член для линейной интерполяции (между 0 и 1):

\begin{equation*}
    R_1(x) = \frac{f''(\xi)}{2}(x-0)(x-1).
\end{equation*}

Т.к. $f''(x) = 24x$, при некотором $\xi \in (0,1)$, оценим примерно $\xi = 0.5$:

\begin{equation*}
    f''(0.5) = 24 \cdot 0.5 = 12.
\end{equation*}

\begin{equation*}
    R_1(0.5) = \frac{12}{2}(0.5 - 0)(0.5 - 1) = 6 \cdot 0.5 \cdot (-0.5) = 6 \cdot (-0.25) = -1.5.
\end{equation*}

По модулю:

\begin{equation*}
    |R_1(0.5)| = 1.5.
\end{equation*}

Совпадение с $\sigma_1$ полное.

#### Квадратичная интерполяция:

Берём три узла: $x = 0$, $1$, $2$:

\begin{equation*}
    f(0)=1, \text{ } f(1)=5, \text{ } f(2)=33.
\end{equation*}

Квадратичный многочлен Лагранжа, упрощённо:

\begin{equation*}
    P_2(x) = 12x^2 - 8x + 1
\end{equation*}

(проверено по узлам).

В точке $x_0 = 0.5$:

\begin{equation*}
    P_2(0.5) = 12(0.25) - 8(0.5) + 1 = 3 - 4 + 1 = 0.
\end{equation*}

Точное:

\begin{equation*}
    f(0.5) = 1.5.
\end{equation*}

Ошибка:

\begin{equation*}
    \sigma_2 = |f(0.5) - P_2(0.5)| = |1.5 - 0| = 1.5.
\end{equation*}

Остаточный член для квадратичной интерполяции:

\begin{equation*}
    R_2(x) = \frac{f^{(3)}(\xi)}{3!}(x-0)(x-1)(x-2),
\end{equation*}

где $f^{(3)}(x)=24$.

\begin{equation*}
    \frac{24}{3!} = \frac{24}{6} = 4.
\end{equation*}

В точке $x = 0.5$:

\begin{equation*}
    (0.5 - 0)(0.5 - 1)(0.5 - 2) = 0.5 \cdot (-0.5) \cdot (-1.5) = 0.5 \cdot (-0.5) = -0.25, \text{ } -0.25 \cdot (-1.5) = 0.375.
\end{equation*}

\begin{equation*}
    R_2(0.5) = 4 \cdot 0.375 = 1.5.
\end{equation*}

По модулю:

\begin{equation*}
    |R_2(0.5)| = 1.5.
\end{equation*}

Совпадает с $\sigma_2$.

#### Кубическая интерполяция:

Берём четыре точки, например $x=0$, $1$, $2$, $3$. Функция сама кубическая, значит интерполяционный многочлен совпадает с ней:

\begin{equation*}
    P_3(x) = 4x^3 + 1.
\end{equation*}

В точке $0.5$:

\begin{equation*}
    P_3(0.5) = f(0.5)=1.5.
\end{equation*}

Ошибка:
\begin{equation*}
\sigma_3 = |f(0.5) - P_3(0.5)| = 0.
\end{equation*}

Остаточный член при кубической интерполяции для кубического полинома равен нулю.

---

#### Вывод для $x_0 = 0.5$:

\begin{equation*}
    \sigma_1=1.5, \text{ } |R_1|=1.5;\quad \sigma_2=1.5, \text{ } |R_2|=1.5; \quad \sigma_3=0.
\end{equation*}

### б)

Берём узлы: $x=2$ и $x=3$:

\begin{equation*}
f(2)=33,\text{ } f(3)=109.
\end{equation*}

\begin{equation*}
P_1(x)=f(2)\frac{x-3}{2-3} + f(3)\frac{x-2}{3-2}=33\frac{x-3}{-1}+109(x-2).
\end{equation*}

\begin{equation*}
P_1(x)=-33(x-3)+109(x-2)= -33x+99+109x-218=76x -119.
\end{equation*}

\begin{equation*}
P_1(2.5)=76\cdot2.5 -119=190-119=71.
\end{equation*}

\begin{equation*}
f(2.5)=4(2.5^3)+1=4(15.625)+1=62.5+1=63.5.
\end{equation*}

Ошибка:
\begin{equation*}
\sigma_1 = |f(2.5)-P_1(2.5)| = |63.5-71|=7.5.
\end{equation*}

\begin{equation*}
R_1(x)=\frac{f''(\xi)}{2}(x-2)(x-3),
\end{equation*}
где $f''(x)=24x$. На интервале (2,3) возьмём $\xi=2.5$:
\begin{equation*}
f''(2.5)=24\cdot2.5=60.
\end{equation*}

\begin{equation*}
R_1(2.5)=\frac{60}{2}(0.5)(-0.5)=30\cdot(-0.25)=-7.5.
\end{equation*}

По модулю:
\begin{equation*}
|R_1(2.5)|=7.5,

\begin{equation*}
L_2(x)=\frac{(x-2)(x-3)}{(4-2)(4-3)}=\frac{(x-2)(x-3)}{2}.

Подставим $x=2.5$:

2.5-3=-0.5,\text{ } 2.5-4=-1.5,\text{ } 2.5-2=0.5.

L_0(2.5)=\frac{(-0.5)(-1.5)}{2}=\frac{0.75}{2}=0.375.

\begin{equation*}
L_1(2.5)=-(0.5\cdot(-1.5))= -(-0.75)=0.75.
\end{equation*}

\begin{equation*}
L_2(2.5)=\frac{(0.5)(-0.5)}{2}=\frac{-0.25}{2}=-0.125.
\end{equation*}

\begin{equation*}
P_2(2.5)=33\cdot0.375 +109\cdot0.75 +257\cdot(-0.125) =12.375+81.75 -32.125 =62.

Точное:
\begin{equation*}
f(2.5)=63.5.
\end{equation*}

Ошибка:
\begin{equation*}
\sigma_2=|63.5-62|=1.5.
\end{equation*}

Остаточный член:
\begin{equation*}
R_2(x)=\frac{f^{(3)}(\xi)}{3!}(x-2)(x-3)(x-4), \quad f^{(3)}(x)=24.

\begin{equation*}
\frac{24}{6}=4.
\end{equation*}

\begin{equation*}
(2.5-2)(2.5-3)(2.5-4)=0.5\cdot(-0.5)\cdot(-1.5)=0.5\cdot(-0.5)=-0.25,\text{ } -0.25\cdot(-1.5)=0.375.

\begin{equation*}
R_2(2.5)=4\cdot0.375=1.5.

По модулю:
\begin{equation*}
|R_2(2.5)|=1.5,

что совпадает с $\sigma_2$.

## 4.1.13. [<img src="https://raw.githubusercontent.com/FortAwesome/Font-Awesome/6.x/svgs/solid/arrow-rotate-left.svg" width="20" height="20" style="filter: invert(21%) sepia(96%) saturate(3035%) hue-rotate(204deg) brightness(99%) contrast(103%);">](#00)

Для заданной таблицы значений построить интерполяционный многочлен, обеспечивающий погрешность интерполяции, соответствующую точности задания таблицы, и вычислить его значение при $x = 0.015$.

| X | $0.0000$ | $0.0100$ | $0.0200$ | $0.0300$ | $0.0400$ | $0.0500$ |
| --- | --- | --- | --- | --- | --- | --- |
| y | $1.0000  | $1.0468$ | $1.0934 $ | $1.1402$ | $1.1632$ | $1.2217$ |

##### Определение шага и узлов интерполяции:

Имеем равномерную сетку по $x$ с шагом $ h = 0.01 $:

\begin{equation*}
x_0 = 0.00, \quad x_1 = 0.01, \quad x_2 = 0.02, \quad x_3 = 0.03, \quad x_4 = 0.04, \quad x_5 = 0.05.
\end{equation*}

Значения функции в узлах:

\begin{equation*}
f(x_0)=f(0.00)=1.0000, \quad f(x_1)=1.0468, \quad f(x_2)=1.0934, \quad f(x_3)=1.1402, \quad f(x_4)=1.1632, \quad f(x_5)=1.2217.
\end{equation*}

##### Таблица конечных разностей:

Построим таблицу конечных разностей $\Delta f$:

Первая разность ($\Delta f$):

\begin{equation*}
\Delta f(x_0) = f(x_1)-f(x_0) = 1.0468 - 1.0000 = 0.0468
\end{equation*}

\begin{equation*}
\Delta f(x_1) = f(x_2)-f(x_1) = 1.0934 - 1.0468 = 0.0466
\end{equation*}

\begin{equation*}
\Delta f(x_2) = f(x_3)-f(x_2) = 1.1402 - 1.0934 = 0.0468
\end{equation*}

\begin{equation*}
\Delta f(x_3) = f(x_4)-f(x_3) = 1.1632 - 1.1402 = 0.0230
\end{equation*}

\begin{equation*}
\Delta f(x_4) = f(x_5)-f(x_4) = 1.2217 - 1.1632 = 0.0585
\end{equation*}

Вторая разность ($\Delta^2 f$):

\begin{equation*}
\Delta^2 f(x_0) = \Delta f(x_1)-\Delta f(x_0) = 0.0466 - 0.0468 = -0.0002
\end{equation*}

\begin{equation*}
\Delta^2 f(x_1) = \Delta f(x_2)-\Delta f(x_1) = 0.0468 - 0.0466 = 0.0002
\end{equation*}

\begin{equation*}
\Delta^2 f(x_2) = \Delta f(x_3)-\Delta f(x_2) = 0.0230 - 0.0468 = -0.0238
\end{equation*}

\begin{equation*}
\Delta^2 f(x_3) = \Delta f(x_4)-\Delta f(x_3) = 0.0585 - 0.0230 = 0.0355
\end{equation*}

Третья разность ($\Delta^3 f$):

\begin{equation*}
\Delta^3 f(x_0) = \Delta^2 f(x_1)-\Delta^2 f(x_0) = 0.0002 - (-0.0002) = 0.0004
\end{equation*}

\begin{equation*}
\Delta^3 f(x_1) = \Delta^2 f(x_2)-\Delta^2 f(x_1) = -0.0238 - 0.0002 = -0.0240

\Delta^3 f(x_2) = \Delta^2 f(x_3)-\Delta^2 f(x_2) = 0.0355 - (-0.0238) = 0.0593

\end{equation*}

Итого, у нас есть набор начальных разностей:

\begin{equation*}
f(x_0)=1.0000, \quad \Delta f(x_0)=0.0468, \quad \Delta^2 f(x_0)=-0.0002, \quad \Delta^3 f(x_0)=0.0004, \quad \Delta^4 f(x_0)=-0.0244, \quad \Delta^5 f(x_0)=0.1077.
\end{equation*}

##### Формула Ньютона для равностоящих узлов:

Интерполяционный многочлен Ньютона для равномерных узлов имеет вид:

\begin{equation*}
P_n(x) = f(x_0) + t \Delta f(x_0) + \frac{t(t-1)}{2!}\Delta^2 f(x_0) + \frac{t(t-1)(t-2)}{3!}\Delta^3 f(x_0) + \frac{t(t-1)(t-2)(t-3)}{4!}\Delta^4 f(x_0) + \frac{t(t-1)(t-2)(t-3)(t-4)}{5!}\Delta^5 f(x_0)
\end{equation*}

Где $ t = \frac{x - x_0}{h} $. В нашем случае $ x_0=0 $, $ h=0.01 $. Для $ x=0.015 $:

\begin{equation*}
t = \frac{0.015 - 0.00}{0.01} = 1.5.
\end{equation*}

Подставим $ t=1.5 $:

- $ t=1.5 $
- $ t(t-1)=1.5 \times 0.5=0.75 $
- $ t(t-1)(t-2)=0.75 \times (-0.5) = -0.375 $
- $ t(t-1)(t-2)(t-3) = (-0.375)\times(-1.5)=0.5625 $
- $ t(t-1)(t-2)(t-3)(t-4)=0.5625 \times (-2.5) = -1.40625 $

Теперь подставим в формулу Ньютона:

\begin{equation*}
P_5(0.015)=f(x_0) + t\Delta f(x_0) + \frac{t(t-1)}{2!}\Delta^2 f(x_0) + \frac{t(t-1)(t-2)}{3!}\Delta^3 f(x_0) + \frac{t(t-1)(t-2)(t-3)}{4!}\Delta^4 f(x_0) + \frac{t(t-1)(t-2)(t-3)(t-4)}{5!}\Delta^5 f(x_0).
\end{equation*}

Подставим численные значения:

- $ f(x_0)=1.0000 $
- $ \Delta f(x_0)=0.0468 $
- $ \Delta^2 f(x_0)=-0.0002 $
- $ \Delta^3 f(x_0)=0.0004 $
- $ \Delta^4 f(x_0)=-0.0244 $
- $ \Delta^5 f(x_0)=0.1077 $

И учитываем факториалы: $2!=2, \ 3!=6, \ 4!=24, \ 5!=120.$

\begin{equation*}
P_5(0.015) = 1.0000 + (1.5)(0.0468) + \frac{0.75}{2}(-0.0002) + \frac{-0.375}{6}(0.0004) + \frac{0.5625}{24}(-0.0244) + \frac{-1.40625}{120}(0.1077).
\end{equation*}

Выполним по шагам:

1. $ (1.5)(0.0468)=0.0702 $
2. $\frac{0.75}{2}=0.375,\; 0.375 \times (-0.0002)=-0.000075 $
3. $\frac{-0.375}{6}=-0.0625,\; -0.0625 \times 0.0004=-0.000025 $
4. $\frac{0.5625}{24}=0.0234375,\; 0.0234375 \times (-0.0244)=-0.0005715 $
5. $\frac{-1.40625}{120}=-0.01171875,\; -0.01171875 \times 0.1077 \approx -0.0012617$

Сложим всё:

\begin{equation*}
1.0000 + 0.0702 -0.000075 -0.000025 -0.0005715 -0.0012617
\end{equation*}

Пошагово:

- $1.0000 + 0.0702 = 1.0702$
- $1.0702 - 0.000075 = 1.070125$
- $1.070125 - 0.000025 = 1.0701$ (округленно)
- $1.0701 - 0.0005715 = 1.0695285$
- $1.0695285 - 0.0012617 \approx 1.0682668$

Округляя до точности, сопоставимой с исходной таблицей (4 знака после запятой):

\begin{equation*}
P_5(0.015) \approx 1.0683.
\end{equation*}

**Ответ:** Значение интерполяционного многочлена при $ x=0.015 $ примерно равно $ 1.0683 $.

## 4.1.17. [<img src="https://raw.githubusercontent.com/FortAwesome/Font-Awesome/6.x/svgs/solid/arrow-rotate-left.svg" width="20" height="20" style="filter: invert(21%) sepia(96%) saturate(3035%) hue-rotate(204deg) brightness(99%) contrast(103%);">](#00)

Построить дважды непрерывно дифференцируемый кубический сплайн для таблицы значений $y(x)$. Принять $y'(-1) = 4$, $y'(1) = -2$.

| x | $-1$ | $0$ | $1$ |
| --- | --- | --- | --- |
| y | $-1$  | $1$ | $1$ |

#### Зададим общую форму сплайна:

Разобьем отрезок $[-1,1]$ на два подынтервала: $[-1,0]$ и $[0,1]$.

На каждом подынтервале определим кубический полином:

1. На $[-1,0]$: 
\begin{equation*}
S_1(x) = a_1 x^3 + b_1 x^2 + c_1 x + d_1.
\end{equation*}

2. На $[0,1]$:
\begin{equation*}
S_2(x) = a_2 x^3 + b_2 x^2 + c_2 x + d_2.
\end{equation*}

Итого у нас 8 неизвестных коэффициентов: $a_1, b_1, c_1, d_1, a_2, b_2, c_2, d_2.$

#### Записываем условия:

1. Интерполяция значений в узлах:
   - При $x=-1$: $S_1(-1)=y(-1)=-1.$
   - При $x=0$: $S_1(0)=y(0)=1$ и одновременно $S_2(0)=y(0)=1$ (обеспечивает непрерывность в 0).
   - При $x=1$: $S_2(1)=y(1)=1.$

2. Граничные условия на производные:
   - $S_1'(-1) = y'(-1)=4.$
   - $S_2'(1) = y'(1)=-2.$

3. Условия гладкости в точке стыка $x=0$:
   - Непрерывность первой производной: $S_1'(0)=S_2'(0)$.
   - Непрерывность второй производной: $S_1''(0)=S_2''(0)$.

Таким образом, у нас будет 8 уравнений для 8 неизвестных.

#### Запись уравнений:

Рассмотрим сначала уравнения для первого интервала:

- Значение в $-1$:
  \begin{equation*}
  S_1(-1)=a_1(-1)^3+b_1(-1)^2+c_1(-1)+d_1=-a_1+b_1 - c_1 + d_1 = -1.
  \end{equation*}

- Значение в $0$:
  \begin{equation*}
  S_1(0)=d_1=1 \implies d_1=1.
  \end{equation*}

Подставим $d_1=1$ в первое уравнение:

\begin{equation*}
 -a_1 + b_1 - c_1 + 1 = -1 \implies -a_1 + b_1 - c_1 = -2.
\end{equation*}

- Производная на $[-1,0]$:
  \begin{equation*}
  S_1'(x)=3a_1x^2 + 2b_1x + c_1.
  \end{equation*}
  При $x=-1$:
  \begin{equation*}

## 4.2.1. [<img src="https://raw.githubusercontent.com/FortAwesome/Font-Awesome/6.x/svgs/solid/arrow-rotate-left.svg" width="20" height="20" style="filter: invert(21%) sepia(96%) saturate(3035%) hue-rotate(204deg) brightness(99%) contrast(103%);">](#00)

Построить наилучшее среднеквадратичное приближение в виде $y = a + b \lg(x)$, где $a$, $b = \text{const}$, для таблично заданной функции:

| x | $10$ | $1000$ | $100000$ |
| --- | --- | --- | --- |
| y | $10$  | $100$ | $1000$ |

Введём обозначение:
\begin{equation*}
L = \lg(x).
\end{equation*}

Для заданных точек найдём $\lg(x)$:

1. $\lg(10) = 1$
2. $\lg(1000) = \lg(10^3) = 3$
3. $\lg(100000) = \lg(10^5) = 5.$

Таким образом, исходные точки $(x_i, y_i)$ перейдут в точки $(L_i, y_i)$:

\begin{equation*}
(L_1, y_1) = (1, 10), \quad (L_2, y_2) = (3, 100), \quad (L_3, y_3) = (5, 1000).
\end{equation*}

#### Метод наименьших квадратов:

Мы подбираем прямую:
\begin{equation*}
y = a + bL.
\end{equation*}

Неизвестны $a$ и $b$.

Запишем нормальные уравнения для МНК для набора из $n=3$ точек:

\begin{equation*}
\sum_{i=1}^n y_i = a\sum_{i=1}^n 1 + b \sum_{i=1}^n L_i,
\end{equation*}

\begin{equation*}
\sum_{i=1}^n L_i y_i = a\sum_{i=1}^n L_i + b \sum_{i=1}^n L_i^2.
\end{equation*}

Подставим наши данные:

1. Суммируем значения $L_i$:
\begin{equation*}
\sum L_i = 1+3+5 = 9.
\end{equation*}

1. Суммируем значения $y_i$:
\begin{equation*}
\sum y_i = 10 + 100 + 1000 = 1110.
\end{equation*}

1. Суммируем квадраты $L_i^2$:
\begin{equation*}
L_1^2=1^2=1, \; L_2^2=3^2=9, \; L_3^2=5^2=25,
\end{equation*}

\begin{equation*}
\sum L_i^2 = 1+9+25=35.
\end{equation*}

1. Суммируем произведения $L_i y_i$:
\begin{equation*}
L_1 y_1 = 1 \cdot 10 =10,\quad L_2 y_2 = 3 \cdot 100=300,\quad L_3 y_3=5\cdot 1000=5000,
\end{equation*}

\begin{equation*}
\sum L_i y_i = 10+300+5000=5310.
\end{equation*}

Теперь подставляем в систему:

Из первого уравнения:
\begin{equation*}
\sum y_i = 1110 = a \cdot 3 + b \cdot 9,
\end{equation*}
так как $\sum 1 = 3$ (три точки), и $\sum L_i=9$.

Получаем:
\begin{equation*}
3a + 9b = 1110. \quad (1)
\end{equation*}

Из второго уравнения:
\begin{equation*}
\sum L_i y_i = 5310 = a \sum L_i + b \sum L_i^2 = a \cdot 9 + b \cdot 35.
\end{equation*}

Имеем:
\begin{equation*}
9a + 35b = 5310. \quad (2)
\end{equation*}

Решаем систему линейных уравнений

Из уравнения (1):
\begin{equation*}
3a + 9b=1110 \implies a+3b=370 \implies a=370-3b.
\end{equation*}

Подставим $a=370-3b$ в (2):
\begin{equation*}
9(370-3b)+35b=5310.
\end{equation*}

Раскроем скобки:
\begin{equation*}
3330 -27b +35b=5310,
\end{equation*}

\begin{equation*}
3330 +8b=5310,
\end{equation*}

\begin{equation*}
8b=5310-3330=1980,
\end{equation*}

\begin{equation*}
b=1980/8=247.5.
\end{equation*}

Теперь находим $a$:
\begin{equation*}
a=370-3(247.5)=370-742.5=-372.5.
\end{equation*}

#### Итог:

Мы получили:
\begin{equation*}
a=-372.5, \quad b=247.5.
\end{equation*}

Следовательно, наилучшее среднеквадратичное приближение в искомом виде:
\begin{equation*}
y = a + b \lg(x) = -372.5 + 247.5 \lg(x).
\end{equation*}

## 4.3.3. [<img src="https://raw.githubusercontent.com/FortAwesome/Font-Awesome/6.x/svgs/solid/arrow-rotate-left.svg" width="20" height="20" style="filter: invert(21%) sepia(96%) saturate(3035%) hue-rotate(204deg) brightness(99%) contrast(103%);">](#00)

Задача сведётся к задаче равносильной приближению нуля данным семейством многочленов в равномерной норме. P(0) = -M,\quad P(1) = M,\quad P(2) = -M.
\end{equation*}

\begin{equation*}
P(0) = a_0 = -M.
\end{equation*}

\begin{equation*}
P(1) = a_2 \cdot 1^2 + 1 + a_0 = a_2 + 1 + (-M) = M.
\end{equation*}
Отсюда
\begin{equation*}
a_2 + 1 - M = M \implies a_2 + 1 = 2M \implies a_2 = 2M - 1.
\end{equation*}

\begin{equation*}
P(2) = a_2 \cdot 2^2 + 2 + a_0 = 4a_2 + 2 + (-M).
\end{equation*}
По условию это должно быть равно &-M&:
\begin{equation*}
4a_2 + 2 - M = -M \implies 4a_2 + 2 = 0 \implies 4a_2 = -2.
\end{equation*}

\begin{equation*}
a_2 = 2M - 1.
\end{equation*}

\begin{equation*}
4(2M - 1) = -2 \implies 8M - 4 = -2 \implies 8M = 2 \implies M = \frac{1}{4}.
\end{equation*}

\begin{equation*}
a_0 = -M = -\frac{1}{4}, \quad a_2 = 2M -1 = \frac{1}{2} -1 = -\frac{1}{2}.
\end{equation*}

\begin{equation*}
P(x) = -\frac{1}{2}x^2 + x - \frac{1}{4}.
\end{equation*}

P(0) = -\frac{1}{4}, \quad P(1) = -\frac{1}{2} + 1 - \frac{1}{4} = -\frac{2}{4} + \frac{4}{4} - \frac{1}{4} = \frac{1}{4}, \quad P(2) = -\frac{1}{2}\cdot 4 + 2 - \frac{1}{4} = -2 + 2 - \frac{1}{4} = -\frac{1}{4}.
\end{equation*}

**Ответ:**
Наименее уклоняющимся от нуля многочленом является
\begin{equation*}
P(x) = -\frac{1}{2}x^2 + x - \frac{1}{4}.
\end{equation*}

## 4.3.9. [<img src="https://raw.githubusercontent.com/FortAwesome/Font-Awesome/6.x/svgs/solid/arrow-rotate-left.svg" width="20" height="20" style="filter: invert(21%) sepia(96%) saturate(3035%) hue-rotate(204deg) brightness(99%) contrast(103%);">](#00)

Построить сетку из пяти точек, минимизирующую оценку остаточного члена полиномиальной интерполяции на отрезке $[0; \text{ } \pi]$ и оценить максимальную погрешность интерполяции функции $\sin (x)$ на этом отрезке.

Для наилучшего равномерного приближения функции многочленами интерполяции обычно используют так называемые **чебышёвские узлы**, поскольку они минимизируют эффект вызванный феноменом Рунге и тем самым дают наименьшую верхнюю оценку погрешности среди всех равномерных расположений точек. Хотя чебышёвская сетка прямо оптимизирует не всегда именно интерполяционную погрешность для произвольной функции, для гладких функций вроде $\sin(x)$ на отрезке $[0,\pi]$ она практически обеспечивает близкие к минимаксимальным результаты.

Мы хотим построить интерполяционный многочлен четвёртой степени (то есть 5 точек). Чебышёвские узлы для интервала $[-1,1]$ задаются формулой:
\begin{equation*}
t_k = \cos\left(\frac{(2k+1)\pi}{2(n+1)}\right),\quad k=0,1,\ldots,n,
\end{equation*}
где $n=4$ для 5 точек. Таким образом,
\begin{equation*}
t_0 = \cos\left(\frac{\pi}{10}\right),\; t_1 = \cos\left(\frac{3\pi}{10}\right),\; t_2 = \cos\left(\frac{5\pi}{10}\right),\; t_3 = \cos\left(\frac{7\pi}{10}\right),\; t_4 = \cos\left(\frac{9\pi}{10}\right).
\end{equation*}

Посчитаем приближённые значения:
- $\cos(\pi/10) \approx 0.951056$
- $\cos(3\pi/10) = \cos(54^\circ) \approx 0.587785$
- $\cos(\pi/2) = 0$
- $\cos(7\pi/10) = \cos(126^\circ) = -\cos(54^\circ) \approx -0.587785$
- $\cos(9\pi/10) = \cos(162^\circ) = -\cos(18^\circ) \approx -0.951056$

Таким образом,
\begin{equation*}
t_0 \approx 0.951056,\;
t_1 \approx 0.587785,\;
t_2 = 0,\;
t_3 \approx -0.587785,\;
t_4 \approx -0.951056.
\end{equation*}

Теперь нам нужно перенести эти точки с $[-1,1]$ на $[0,\pi]$. Линейное отображение $[-1,1]\to[0,\pi]$ задаётся формулой:
\begin{equation*}
x = \frac{\pi}{2}(t+1).
\end{equation*}

Подставим $t_k$:

- $x_0 = \frac{\pi}{2}(0.951056+1) = \frac{\pi}{2}\cdot1.951056 \approx 3.065$
- $x_1 = \frac{\pi}{2}(0.587785+1) = \frac{\pi}{2}\cdot1.587785 \approx 2.493$
- $x_2 = \frac{\pi}{2}(0+1) = \frac{\pi}{2} \approx 1.5708$
- $x_3 = \frac{\pi}{2}( -0.587785 +1) = \frac{\pi}{2}\cdot0.412215 \approx 0.6479$
- $x_4 = \frac{\pi}{2}( -0.951056 +1) = \frac{\pi}{2}\cdot0.048944 \approx 0.0769$

Упорядочим по возрастанию:
\begin{equation*}
x_4 \approx 0.0769,\; x_3 \approx 0.6479,\; x_2 \approx 1.5708,\; x_1 \approx 2.493,\; x_0 \approx 3.065.
\end{equation*}

Это чебышёвская сетка из пяти точек на $[0,\pi]$.

#### Оценка максимальной погрешности:

Интерполяционная формула ошибки для интерполяции функции $ f(x) $ в точках $ x_0, \ldots, x_n $ многочленом степени $n$ имеет вид:
\begin{equation*}
R_n(x) = \frac{f^{(n+1)}(\xi)}{(n+1)!}(x-x_0)(x-x_1)\cdots(x-x_n)
\end{equation*}
для некоторого $\xi \in [0,\pi]$.

Здесь $n=4$. Для $f(x)=\sin(x)$:
\begin{equation*}
f'(x)=\cos(x),\; f''(x)=-\sin(x),\; f'''(x)=-\cos(x),\; f^{(4)}(x)=\sin(x),\; f^{(5)}(x)=\cos(x).
\end{equation*}

Следовательно, $ f^{(5)}(\xi)=\cos(\xi) $, и максимальный модуль пятой производной на $[0,\pi]$ равен 1. Значит
\begin{equation*}
|R_4(x)| \le \frac{|(x-x_0)(x-x_1)(x-x_2)(x-x_3)(x-x_4)|}{5!} = \frac{|(x-x_0)(x-x_1)(x-x_2)(x-x_3)(x-x_4)|}{120}.
\end{equation*}

Чтобы оценить максимум, рассмотрим величину $|(x-x_0)(x-x_1)(x-x_2)(x-x_3)(x-x_4)|$ на $[0,\pi]$. При чебышёвских узлах этот максимум минимален среди всех возможных сеток. Приближённо оценим:

На концах интервала, например в точке 0:
\begin{equation*}
|W(0)| = |(-x_0)(-x_1)(-x_2)(-x_3)(-x_4)| \approx 3.065 \cdot 2.493 \cdot 1.5708 \cdot 0.6479 \cdot 0.0769 \approx 0.6.
\end{equation*}

Аналогичное значение будет и в окрестности $\pi$. Поскольку чебышёвская сетка симметрична, можно считать, что максимум порядка 0.6 - 0.7. Возьмём оценку около 0.6.

Тогда 
\begin{equation*}
|R_4(x)| \le \frac{0.6}{120} = 0.005.
\end{equation*}

Таким образом, максимальная погрешность интерполяции $\sin(x)$ многочленом степени 4 при выборе чебышёвских узлов на $[0,\pi]$ будет порядка 0.005 (около 0.5%).

**Ответ:**

1. Наилучшим выбором сетки из пяти точек, минимизирующим оценку ошибки, являются чебышёвские узлы, переведённые с $[-1,1]$ на $[0,\pi]$:
\begin{equation*}
x_4 \approx 0.0769,\; x_3 \approx 0.6479,\; x_2 = \frac{\pi}{2}\approx 1.5708,\; x_1 \approx 2.493,\; x_0 \approx 3.065.
\end{equation*}

1. Максимальная погрешность интерполяции $\sin(x)$ на $[0,\pi]$ многочленом, построенным в этих узлах, не превысит примерно 0.005.

    y_{n+3} - 3y_{n+2} + 9y_{n+1} - 27y_{n} = 0
\end{equation*}

Подставим решение вида $y_n = r^n$:

\begin{equation*}
    r^{n+3} - 3r^{n+2} + 9r^{n+1} - 27r^n = 0.
\end{equation*}

Разделив на $r^n$ (предполагая $r \neq 0$), получаем характеристическое уравнение:

\begin{equation*}
    r^3 - 3r^2 + 9r - 27 = 0.
\end{equation*}

Проверим, является ли $r = 3$ корнем:

\begin{equation*}
    3^3 - 3 \cdot 3^2 + 9 \cdot 3 - 27 = 27 - 27 + 27 - 27 = 0.
\end{equation*}

Значит, $r = 3$ – корень. Теперь выполним факторизацию:

\begin{equation*}
    r^3 - 3r^2 + 9r - 27 = (r - 3)(r^2 + 9).
\end{equation*}

Уравнение $r^2 + 9 = 0$ даёт ещё два корня:

\begin{equation*}
    r^2 = -9 \implies r = \pm 3i.
\end{equation*}

Таким образом, корни характеристического уравнения:

\begin{equation*}
    r_1 = 3, \quad r_2 = 3i, \quad r_3 = -3i.
\end{equation*}

Общее решение линейного однородного разностного уравнения - это линейная комбинация членов вида $r^n$ для каждого корня $r$.

Для действительного решения, учитывая комплексные корни, удобнее перейти к их тригонометрическому представлению. Заметим, что корни $3i$ и $-3i$ имеют модуль $3$ и аргумент $\pm \frac{\pi}{2}$. Решение для комплексно-сопряжённых корней $3i$ и $-3i$ можно записать в форме, выражающей реальную часть:

\begin{equation*}
    y_n = A \cdot 3^n + 3^n (B \cos(\tfrac{\pi}{2} n) + C \sin(\tfrac{\pi}{2} n)),
\end{equation*}

где $A, B, C$ – произвольные постоянные, определяемые начальными условиями.

Итого:

\begin{equation*}
    y_n = 3^n \left( A + B\cos\left(\frac{\pi n}{2}\right) + C\sin\left(\frac{\pi n}{2}\right) \right).
\end{equation*}

\begin{equation*}

\end{equation*}

\begin{equation*}

\end{equation*}

Решаем квадратное уравнение, получаем два корня:

\begin{equation*}

\end{equation*}

Общее решение однородного уравнения:

\begin{equation*}
   y_n^{(h)} = A \cdot 1^n + B \left(\frac{1}{3}\right)^n = A + B\left(\frac{1}{3}\right)^n,
\end{equation*}

где $A$ и $B$ – произвольные постоянные.

Правая часть уравнения – линейная функция от $n$: $2 a n h^2$.

Обычно для правой части вида полинома от $n$ мы подбираем частное решение в виде полинома той же степени. Однако среди корней характеристического уравнения есть корень $r=1$. Это означает «резонанс»: для линейного по $n$ свободного члена нужно взять частное решение на один порядок выше. Изначально для линейной правой части $2 a n h^2$ мы бы попробовали $y_n^{(p)} = Cn + D$. Но из-за присутствия корня $r=1$ в характер. ур-нии нам нужно повысить степень и взять:

\begin{equation*}
   y_n^{(p)} = E n^2 + F n.
\end{equation*}

Подставим $y_n^{(p)}$ в левую часть уравнения.

Вычислим сдвиги:

\begin{equation*}
   y_{n-1}^{(p)} = E(n-1)^2 + F(n-1) = E(n^2 - 2n +1) + F(n-1),
\end{equation*}

\begin{equation*}

\end{equation*}

\end{equation*}

\begin{equation*}

\end{equation*}

Распишем и соберём по степеням:

1. Коэффициенты при $n^2$:
\begin{equation*}

\end{equation*}

1. Коэффициенты при $n$:

\end{equation*}

\end{equation*}

\end{equation*}

- При $n$: $4 E = 2 a h^2 \implies E = \frac{a h^2}{2}$.
- Свободный член: $2 F = 0 \implies F=0$.

Таким образом, частное решение:
\begin{equation*}

\end{equation*}

\begin{equation*}

\end{equation*}

## 6.2.1 [<img src="https://raw.githubusercontent.com/FortAwesome/Font-Awesome/6.x/svgs/solid/arrow-rotate-left.svg" width="20" height="20" style="filter: invert(21%) sepia(96%) saturate(3035%) hue-rotate(204deg) brightness(99%) contrast(103%);">](#00)

С каким порядком на сетке $\overline{D_h} = \{ x_l = lh, \; l = \overline{0, \; L}, \; h = 1/L \{$ разностная схема a) аппроксимирует дифференциальную задачу б) на решении $y$?

а) 

\begin{equation*}
    \left . \begin{split}
        &\frac{y_{l+1} - 2y_{l} + y_{l-1}}{h^2} - y_l = 0, \quad l = \overline{1, \; L-1},\\
        &y_0 = 1, \quad y_1 = 1 + h + \frac{h^2}{2}
    \end{split} \right \}
\end{equation*}

б)

\begin{equation*}
    \left . \begin{split}
        &\frac{d^2 y}{dx^2} - y = 0, \quad 0 \leq x \leq 1,\\
        &y(0) = 1, \quad y'_x(0) = 1.
    \end{split} \right \}
\end{equation*}

**Дифференциальная задача б):**

\begin{equation*}
\begin{cases}
y''(x) - y(x) = 0, \quad x \in [0,1],\\
y(0) = 1, \\
y'_x(0) = 1.
\end{cases}
\end{equation*}

Её точное решение:

\begin{equation*}
y(x) = e^x.
\end{equation*}

Проверим это:

\begin{equation*}
y''(x) = e^x,\; y(x) = e^x \implies y''(x)-y(x)=0.
\end{equation*}

Начальные условия:
\begin{equation*}
y(0)=e^0=1, \quad y'(x)=e^x \implies y'(0)=1.
\end{equation*}

**Разностная схема а):**

Зададим сетку $\overline{D_h}=\{x_l = l h,\; l=\overline{0,L},\; h=\tfrac{1}{L}\}$. Рассмотрим следующую схему:

\begin{equation*}
\frac{y_{l+1}-2y_l+y_{l-1}}{h^2} - y_l = 0, \quad l=\overline{1,L-1},
\end{equation*}

при граничных условиях:

\begin{equation*}
y_0=1,\quad y_1 = 1 + h + \frac{h^2}{2}.
\end{equation*}

Обозначим через $[y]_l = y(x_l)=e^{x_l}=e^{lh}$ точное решение на узлах сетки. Рассмотрим вектор  
\begin{equation*}
Y = ([y]_0,\; [y]_1,\; \ldots,\; [y]_L).
\end{equation*}

Подставим этот вектор в разностную схему и определим вектор неувязки $\delta F(Y)$.

Вектор неувязки $\delta F(Y)$ определяется как значения левой части разностной схемы минус правую часть уравнения, при подстановке точного решения. В нашем случае правая часть равна нулю (однородное уравнение). Также учтём условия на концах:

1. Для внутренних узлов $l=\overline{1,L-1}$:

\begin{equation*}
\delta F_l(Y) = \frac{[y]_{l+1}-2[y]_l+[y]_{l-1}}{h^2} - [y]_l.
\end{equation*}

2. Для граничных условий определим невязку как:

\begin{equation*}
\delta F_0(Y) = [y]_0 - 1, \quad \delta F_1(Y) = [y]_1 - \left(1 + h + \frac{h^2}{2}\right).
\end{equation*}

(Здесь мы рассматриваем точку $l=1$ как условие на второй узел, заданное схемой. Можно было бы рассмотреть систему уравнений в векторном виде, включив начальные условия как отдельные уравнения.)

Для функции $y(x)=e^x$ известна классическая аппроксимация второй производной центральной разностной формулой:

\begin{equation*}
y''(x_l) = \frac{y(x_l+h)-2y(x_l)+y(x_l-h)}{h^2} - C h^2 + \mathcal{O}(h^4),
\end{equation*}

где $C$ – некоторая константа, зависящая от высших производных $y$. Точнее, разложение Тейлора даёт:

\begin{equation*}
y(x_l+h) = y(x_l) + y'(x_l)h + \frac{y''(x_l)h^2}{2} + \frac{y^{(3)}(x_l)h^3}{6} + \mathcal{O}(h^4),
\end{equation*}

\begin{equation*}
y(x_l-h) = y(x_l) - y'(x_l)h + \frac{y''(x_l)h^2}{2} - \frac{y^{(3)}(x_l)h^3}{6} + \mathcal{O}(h^4).
\end{equation*}

Складывая их:

\begin{equation*}
y(x_l+h) - 2y(x_l) + y(x_l-h) = y''(x_l)h^2 + \frac{y^{(4)}(x_l)}{12}h^4 + \mathcal{O}(h^5).
\end{equation*}

Отсюда следует, что:

\begin{equation*}
\frac{y(x_l+h)-2y(x_l)+y(x_l-h)}{h^2} = y''(x_l) + \mathcal{O}(h^2).
\end{equation*}

Так как $y''(x)=y(x)$ для $y(x)=e^x$, мы имеем:

\begin{equation*}
\frac{[y]_{l+1}-2[y]_l+[y]_{l-1}}{h^2} = [y]_l + \mathcal{O}(h^2).
\end{equation*}

Подставляя это в определение $\delta F_l(Y)$:

\begin{equation*}
\delta F_l(Y) = ([y]_l + O(h^2)) - [y]_l = \mathcal{O}(h^2).
\end{equation*}

Для всех внутренних узлов $l=\overline{1,L-1}$ невязка порядка $\mathcal{O}(h^2)$.

- При $l=0$:
\begin{equation*}
[y]_0 = e^{0}=1, \quad \delta F_0(Y) = [y]_0 - 1 = 0.
\end{equation*}

- При $l=1$:
\begin{equation*}
[y]_1 = e^h = 1 + h + \frac{h^2}{2} + \frac{h^3}{6} + \mathcal{O}(h^4).
\end{equation*}

Условие схемы:
\begin{equation*}
y_1 = 1+h+\frac{h^2}{2}.
\end{equation*}

Невязка:
\begin{equation*}
\delta F_1(Y) = [y]_1 - \left(1+h+\frac{h^2}{2}\right) = \frac{h^3}{6}+O(h^4) = \mathcal{O}(h^3).
\end{equation*}

Заметим, что граничное условие в точке $x_1=h$ аппроксимирует точное решение с ещё более высоким порядком точности (погрешность $\mathcal{O}(h^3)$).

Соберём оценку по всей сетке. Вектор $\delta F(Y)$ имеет компоненты:

- $\delta F_0(Y)=0$,
- $\delta F_1(Y)=\mathcal{O}(h^3)$,
- $\delta F_l(Y)=\mathcal{O}(h^2)$ для $l=\overline{2,L-1}$.

Следовательно, если взять норму невязки, например максимум по индексам $l$:

\begin{equation*}
\|\delta F(Y)\|_{\infty} = \mathcal{O}(h^2).
\end{equation*}

Это означает, что разностная схема аппроксимирует исходную дифференциальную задачу с порядком ошибки $\mathcal{O}(h^2)$ относительно точного решения $y(x)=e^x$.

    \begin{split}
        &-(l + 1) (A(l + 1) + B) + 4l(Al + B) - 3(l - 1)(A(l - 1) + B) = 2ah^2l + 2hb\\
        &A(-l^2 - 2l - 1 + 4l^2 - 3l^2 + 6l - 3) + B(-(l + 1) + 4l - 3(l - 1)) = 2ah^2l + 2hb\\
        & \Rightarrow \quad \begin{cases}
            Al = 2ah^2l\\
            2B - 4A = 2hb
        \end{cases} \quad \Rightarrow \quad \begin{cases}
            A = \frac{ah^2}{2}\\
            B = bh + ah^2
        \end{cases}
    \end{split}
\end{equation*}

Тогда общее решение без учёта начальных условий есть

\begin{equation*}
    y_l = C_1 + C_2 \cdot 3^l + l \left ( \frac{ah^2l}{2} + (bh + ah^2) \right )
\end{equation*}

Начальные условия:

\begin{equation*}
    \begin{cases}
        y_0 = 0 \quad \Rightarrow \quad C_1 = - \frac{3ah^2}{4}\\
        y_1 = C_1 + 3C_2 + \dfrac{3ah^2}{2} + bh = bh \quad \Rightarrow \quad C_2 = \dfrac{3ah^2}{4}
    \end{cases}
\end{equation*}

Итого:

\begin{equation*}
    y_l  = \dfrac{3ah^2}{4} - \dfrac{3ah^2}{4} \cdot 3^l + \frac{ah^2l}{2} + hl(b + ah).
\end{equation*}

Наконец расматриваем прямую сходимость (берём первую норму):

\begin{equation*}
    \begin{split}
        || [y]_l - y_l || &= \left| \left| \frac{ah^2l^2}{2} - bhl - \frac{3ah^2}{4} + \frac{3ah^2}{4} \cdot 3^l - \frac{ah^2l^2}{2} - bhl - ah^l \right| \right| =\\
        & = \lim \limits_{\substack{h \to 0 \\ L \to \infty\\ hL = 1}} \max \limits_l \left | \frac{3ah^2}{4} \cdot 3^l - \frac{3ah^2}{4} - ah^2l \right | = \infty
    \end{split}
\end{equation*}

## 6.3.6. [<img src="https://raw.githubusercontent.com/FortAwesome/Font-Awesome/6.x/svgs/solid/arrow-rotate-left.svg" width="20" height="20" style="filter: invert(21%) sepia(96%) saturate(3035%) hue-rotate(204deg) brightness(99%) contrast(103%);">](#00)

Предложить явный двухстадийный метод Рунге–Кутты второго порядка точности для решения задачи Коши системы ОДУ на $[0; \; 1]$:

\begin{equation*}
    \left . \begin{split}
        &\dfrac{dx}{dt} = y + 1\\
        &\dfrac{dy}{dt} = x - t\\
        &x(0) = 1, \quad y(0) = 1.
    \end{split} \right \}
\end{equation*}

Рассмотрим систему ОДУ на интервале $[0, 1]$:

\begin{equation*}
\begin{cases}
\dfrac{dx}{dt} = y + 1,\\
\dfrac{dy}{dt} = x - t,\\
x(0)=1,\quad y(0)=1.
\end{cases}
\end{equation*}

Обозначим вектор $\mathbf{Y}(t) = \begin{pmatrix}x(t) \\ y(t)\end{pmatrix}$. Тогда система может быть записана в векторной форме:

\begin{equation*}
\frac{d\mathbf{Y}}{dt} = \mathbf{f}(t, \mathbf{Y}) = \begin{pmatrix}
y+1 \\ x - t
\end{pmatrix}.
\end{equation*}

Один из классических примеров двухстадийного метода второго порядка – метод средней точки (метод РК2 со стадией в центре шага). Для шага интегрирования $h>0$ и при известных $\mathbf{Y}_n \approx \mathbf{Y}(t_n)$ выполняются следующие действия:

1. Вычисляем промежуточный вектор $\mathbf{k}_1$:

\begin{equation*}
\mathbf{k}_1 = \mathbf{f}(t_n, \mathbf{Y}_n).
\end{equation*}

2. Вычисляем $\mathbf{k}_2$, используя информацию о производной в середине шага $(t_n + h/2)$:

\begin{equation*}
\mathbf{k}_2 = \mathbf{f}\left(t_n + \frac{h}{2}, \; \mathbf{Y}_n + \frac{h}{2}\mathbf{k}_1\right).
\end{equation*}

3. Обновляем решение на следующем шаге:

\begin{equation*}
\mathbf{Y}_{n+1} = \mathbf{Y}_n + h\mathbf{k}_2.
\end{equation*}

Данный метод обладает вторым порядком точности и является явным (т.е. не требуется решение уравнений на каждом шаге).

## 6.3.14. [<img src="https://raw.githubusercontent.com/FortAwesome/Font-Awesome/6.x/svgs/solid/arrow-rotate-left.svg" width="20" height="20" style="filter: invert(21%) sepia(96%) saturate(3035%) hue-rotate(204deg) brightness(99%) contrast(103%);">](#00)

Предложить явный двухстадийный метод Рунге–Кутты второго порядка точности для решения задачи Коши ОДУ на $[0; \;1]$:

\begin{equation*}
    \left . \begin{split}
        &\dfrac{d^2y}{dx^2} + (x + 2) \dfrac{dy}{dx} = 0\\
        &y(0) = 0, \quad y'_x(0) = 1.
    \end{split} \right \}
\end{equation*}

Дано уравнение:

\begin{equation*}
\dfrac{d^2y}{dx^2} + (x + 2) \dfrac{dy}{dx} = 0.
\end{equation*}

Введём новую функцию:
\begin{equation*}
z(x) = \dfrac{dy}{dx}.
\end{equation*}

Тогда исходное уравнение можно записать в виде системы:

\begin{equation*}
\begin{cases}
\dfrac{dy}{dx} = z, \\
\dfrac{dz}{dx} = - (x+2) z.
\end{cases}
\end{equation*}

Начальные условия по исходной задаче:
\begin{equation*}
y(0)=0, \quad y'_x(0)=1.
\end{equation*}
Следовательно,
\begin{equation*}
y(0)=0, \quad z(0)=1.
\end{equation*}

Итак, система ОДУ первого порядка на интервале $[0,1]$ имеет вид:

\begin{equation*}
\begin{cases}
y'(x)=z(x), \\
z'(x)=- (x+2) z(x), \\
y(0)=0, \quad z(0)=1.
\end{cases}
\end{equation*}

Обозначим векторное неизвестное $\mathbf{Y}(x) = \begin{pmatrix} y(x) \\ z(x) \end{pmatrix}$. Тогда:

\begin{equation*}
\frac{d\mathbf{Y}}{dx} = \mathbf{f}(x, \mathbf{Y}) = \begin{pmatrix} z \\ - (x+2) z \end{pmatrix}.
\end{equation*}

Один из стандартных двухстадийных методов второго порядка – метод РК2 типа средней точки. При известном $\mathbf{Y}_n \approx \mathbf{Y}(x_n)$ и шаге $h$ он задаётся формулами:

1. Вычисляем первую стадию:
\begin{equation*}
\mathbf{k}_1 = \mathbf{f}(x_n, \mathbf{Y}_n).
\end{equation*}

1. Вычисляем вторую стадию, берём точку в середине шага $(x_n + \frac{h}{2})$:
\begin{equation*}
\mathbf{k}_2 = \mathbf{f}\left(x_n + \frac{h}{2}, \; \mathbf{Y}_n + \frac{h}{2}\mathbf{k}_1\right).
\end{equation*}

1. Новый шаг:
\begin{equation*}
\mathbf{Y}_{n+1} = \mathbf{Y}_n + h \mathbf{k}_2.
\end{equation*}

Подставим конкретную систему:

- Первая стадия:
\begin{equation*}
\mathbf{k}_1 = \begin{pmatrix} z_n \\ -(x_n+2)z_n \end{pmatrix}.
\end{equation*}

- Вторая стадия:
\begin{equation*}
\mathbf{Y}_n + \frac{h}{2}\mathbf{k}_1 = \begin{pmatrix}
y_n + \dfrac{h}{2} z_n \\
z_n + \dfrac{h}{2}\bigl[-(x_n+2)z_n\bigr]
\end{pmatrix} 
= \begin{pmatrix}
y_n + \dfrac{h}{2} z_n \\
z_n \bigl(1 - \dfrac{h}{2}(x_n+2)\bigr)
\end{pmatrix}.
\end{equation*}

\begin{equation*}
x_n+\frac{h}{2} \text{ подставляем в } \mathbf{f}: 
\quad \mathbf{k}_2 = \begin{pmatrix}
z_n\bigl(1 - \dfrac{h}{2}(x_n+2)\bigr) \\
-\bigl(x_n+\dfrac{h}{2}+2\bigr) z_n\bigl(1 - \dfrac{h}{2}(x_n+2)\bigr)
\end{pmatrix}.
\end{equation*}

- Обновление решения:
\begin{equation*}
y_{n+1} = y_n + h \cdot z_n\left(1 - \frac{h}{2}(x_n+2)\right),
\end{equation*}

\begin{equation*}
z_{n+1} = z_n + h \left[ -\left(x_n+\frac{h}{2}+2\right) z_n\left(1 - \frac{h}{2}(x_n+2)\right) \right].
\end{equation*}

Инициируем методом:
\begin{equation*}
y_0 = 0, \quad z_0 = 1.
\end{equation*}

Это даст нам явную схему второго порядка точности (метод РК2 средней точки), которая при $h \to 0$ сходится к точному решению исходной задачи Коши со скоростью $\mathcal{O}(h^2)$.

## 6.4.3. [<img src="https://raw.githubusercontent.com/FortAwesome/Font-Awesome/6.x/svgs/solid/arrow-rotate-left.svg" width="20" height="20" style="filter: invert(21%) sepia(96%) saturate(3035%) hue-rotate(204deg) brightness(99%) contrast(103%);">](#00)

На сетке $\overline{D_h} = \{ x_l: \; x_l = hl, \; hL = 1, \; l = \overline{0, \; L} \}$ предложить экстраполяционную разностную схему Адамса третьего порядка аппроксимации для решения задачи Коши:

\begin{equation*}
    \left . \begin{split}
        &\dfrac{dy}{dx} = 2y + 4x\\
        &y(0) = 0, \quad x \in [0, \; 1]
    \end{split} \right \}
\end{equation*}

Пусть $x_l = l h$ при $l = 0, 1, \dots, L$ и $hL=1$.   
Обозначим $y_l \approx y(x_l)$ и $f_l = f(x_l, y_l) = 2y_l + 4x_l$.

Схема третьего порядка точности Адамса-Башфорта в общем виде имеет вид:

\begin{equation*}
y_{l+1} = y_{l} + \frac{h}{12} \left[ 23 f_l - 16 f_{l-1} + 5 f_{l-2} \right].
\end{equation*}

Данная схема является экстраполяционной, так как вычисление $y_{l+1}$ происходит на основе значений функции $f$ в предыдущих узлах.

1. Имеется начальное условие $y_0 = y(0)=0$.

2. Для начала применения трехшаговой схемы нам потребуются значения $y_1$ и $y_2$, поскольку схема на каждом шаге использует значения $f_{l}, f_{l-1}, f_{l-2}$.

   Для нахождения $y_1, y_2$ можно воспользоваться, например, методом Эйлера или методом Рунге–Кутты 2-го или 4-го порядка на первых шагах, чтобы "разогнать" схему. После того, как $y_0, y_1, y_2$ будут известны, можно использовать приведенную формулу для $l \geq 2$.

3. Начиная с узла $l=2$, для вычисления $y_{3}$, $y_{4}$ и так далее будет использоваться схема Адамса-Башфорта:

\begin{equation*}
y_{l+1} = y_{l} + \frac{h}{12}\bigl[23(2y_l + 4x_l) -16(2y_{l-1} + 4x_{l-1}) + 5(2y_{l-2} + 4x_{l-2})\bigr].
\end{equation*}

Таким образом, на сетке $\overline{D}_h$ предлагаемая экстраполяционная разностная схема Адамса третьего порядка будет иметь вид:

\begin{equation*}
y_{l+1} = y_{l} + \frac{h}{12}\bigl[23f_l -16f_{l-1} + 5f_{l-2}\bigr], \quad \text{где } f_l = 2y_l + 4x_l.
\end{equation*}

Схема имеет третий порядок точности при условии, что начальные значения $y_1, y_2$ получены с не меньшим порядком точности.
